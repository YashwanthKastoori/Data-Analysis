```python
# clarify what is y and what is x label
#y = df['nextMonth_exrate_movement']
#X = df.drop(['nextMonth_exrate_movement'], axis = 1)

X = df[['percentage_rating','oil_future_price','inflation_index','interest_rate','mb/d',
        'months_into_presidency','current_month','oil_production_price_product']]
y = df.iloc[:, 10].values

# divide train test: 80 % - 20 %
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=109)
```

```python
from sklearn.pipeline import Pipeline
```

```python
# Support Vector Machine
# radial basis function 
#initialize model
svm = SVC(kernel='rbf',probability = True,C=1000,class_weight = None,gamma = 0.0000011111777,degree = 3,break_ties = True)
```

```python
#fit model
svm.fit(X_train, y_train.flatten())

normalized_df_svm_pred = svm.predict(X_test)
```

```python
# Optimization of the model

from sklearn.model_selection import RandomizedSearchCV

#class_weight = None
clf_pipe = Pipeline(steps=[('norm', MinMaxScaler()), ('svm', svm)])
param_grid = {'svm__C': np.logspace(-3, 1.2, 20), 
              'svm__gamma' : np.logspace(-3, 1.2, 20), 
              }

# I used randomized search cv to achieve faster results compared to GridSearchCV. Please check 
# RandomizedSearchCV documents for more information.

optimizer = RandomizedSearchCV(clf_pipe, param_distributions=param_grid, 
                               n_iter=40, cv=3, scoring='roc_auc')

optimizer.fit(X_train, y_train)

# Here are the 5 best test_score parameters out of 40 iterations. 

print('Best test score is : {}'.format(optimizer.best_score_))
pd.DataFrame(optimizer.cv_results_).sort_values('mean_test_score', ascending=False).head()
```

```python
# check accuracy: Accuracy: Overall, how often is the classifier correct? Accuracy = (True Pos + True Negative)/total
acc = accuracy_score(y_test, normalized_df_svm_pred)
print(f"The accuracy score for SVM is: {round(acc,3)*100}%")

# f1 score: The F1 score can be interpreted as a weighted average of the precision and recall, where an F1 score reaches its best value at 1 and worst score at 0.
f1 = f1_score(y_test, normalized_df_svm_pred)
print(f"The f1 score for SVM is: {round(f1,3)*100}%")

# recall score: True Positive Rate(Sensitivity or Recall): When it’s actually yes, how often does it predict yes? True Positive Rate = True Positive/actual yes
recall = recall_score(y_test, normalized_df_svm_pred)
print(f"The recall score for SVM is: {round(recall,3)*100}%")
```

```python
# plotting confusion matrix SVM

cnf_matrix_svm = confusion_matrix(y_test, normalized_df_svm_pred)

sns.heatmap(pd.DataFrame(cnf_matrix_svm), annot=True,cmap="Reds" , fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix SVM\n', y=1.1)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,normalized_df_svm_pred))
print(classification_report(y_test,normalized_df_svm_pred))
```

```python
# In order to scale our train and test data, it's a good practice to create a pipeline.
# Each time pipeline fitted with dataset, it's scaled first then fitted with SVC classifier.
# Therefore scaling seperately for each fold will prevent any data leaks during cross-validation.
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score, validation_curve
from sklearn.metrics import roc_auc_score
from sklearn.svm import SVC
cv = 3 # Change this number to increase number of folds. Note that with increasing folds 
       # cross validation will take much more time.

param_name = 'C'
param_range = np.logspace(-2,1.3,10) # Change it to see how model behaves within different C ranges.

train_score, test_score = validation_curve(clf_pipe, X_train, y_train, 
                                           param_name='svm__{}'.format(param_name),
                                           param_range=param_range, cv=cv, scoring='roc_auc') 
# Plotting Results.

max_train = list(map(np.max, train_score))
min_train = list(map(np.min, train_score))
mean_train = list(map(np.mean, train_score))
max_test = list(map(np.max, test_score))
min_test = list(map(np.min, test_score))
mean_test = list(map(np.mean, test_score))

plt.figure(figsize=(8,6))
plt.plot(param_range, mean_train, color='blue', label='train_score')
plt.fill_between(param_range, max_train, min_train, color='blue', alpha=0.2)
plt.plot(param_range, mean_test, color='orange', label='test_score')
plt.fill_between(param_range, max_test, min_test, color='orange', alpha=0.2)
plt.xlabel('Hyper Parameter ({})'.format(param_name))
plt.ylabel('Roc Auc Score')
plt.title('{}-fold Cross Validation Curve'.format(str(cv)))
plt.legend();
```

```python
# Evaluating Model with previously unseen dataset
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, roc_curve, precision_recall_curve
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVC

clf_pipe.fit(X_train, y_train)
fpr, tpr, thresholds0 = roc_curve(y_test, clf_pipe.decision_function(X_test))
precision, recall, thresholds1 = precision_recall_curve(y_test, 
                                            clf_pipe.decision_function(X_test))

threshold_val = 0.9 # Change this number to adjust decision threshold 
current_threshold1 = np.argmin(list(map(lambda x: np.abs(threshold_val - x), thresholds1)))

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(15,7))
plt.subplot(ax1)
plt.plot(fpr, tpr, color='blue')
plt.plot(np.linspace(0,1,2),np.linspace(0,1,2) , '--', color='black')
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.title('ROC Curve')

plt.subplot(ax2)
plt.plot(precision, recall, color='blue')
plt.scatter(precision[current_threshold1], recall[current_threshold1], 
            edgecolors='red', facecolors='none', marker='o', s=200,
            linewidths = 1.5, label='current threshold')
plt.xlabel('precision')
plt.ylabel('recall')
plt.title('Precision Recall Curve')
plt.legend(loc=3);
```
